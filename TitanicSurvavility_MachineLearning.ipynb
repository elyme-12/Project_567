<h1> BASIC OF MACHINE LEARNING </h1>

```python
import pandas as pd 
```

```python
titanic = pd.read_csv('new_dataset/Titanic.csv') 
```

```python
titanic
```

```python
# Drop irrelevant columns
titanic = titanic[["Survived", "Pclass", "Sex", "Age","Embarked"]]

# Drop missing values
titanic = titanic.dropna()

# Encode categorical variable
titanic["Sex"] = titanic["Sex"].map({"male": 0, "female": 1})
titanic["Embarked"] = titanic["Embarked"].map({"C":0, "Q":1, "S":2})
```

```python
titanic
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = titanic[["Pclass", "Sex", "Age", "Embarked"]]
y = titanic["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

<h1> Accuracy Test </h1>

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h3>Accuracy Result</h3>

<h4> When the result of the accuracy is above 0.6 it means the model is picking up patterns it is now <strong> LEARNING </strong></h4>

```python
passengers = [
    {"Name": "Jack", "Sex": "male", "IsBoarded": 1, "Age": 20},
    {"Name": "Rose", "Sex": "female", "IsBoarded": 1, "Age": 18},
    {"Name": "Southampton", "Sex": "male", "IsBoarded": 0, "Age": 30}
]

def predict_survival(passenger):
    if passenger["Sex"] == "female":
        return 1
    elif passenger["Sex"] == "male" and passenger["Age"] < 18:
        return 1
    else:
        return 0

for passenger in passengers:
    result = predict_survival(passenger)
    if result == 1:
        print(passenger["Name"] + " likely to survive")
    else:
        print(passenger["Name"] + " not likely to survive")
```
